```python
matches = driver.find_elements_by_xpath('//div[@class="event__match event__match--static event__match--twoLine"]')
matches2 = driver.find_elements_by_xpath('//div[@class="event__match event__match--static event__match--last event__match--twoLine"]')
matches = matches + matches2
matches_id = []
for i in range(len(matches)):
    matches_id.append(matches[i].get_attribute('id'))
    
matches_id2 = []
for item in matches_id:
    matches_id2.append(item.split('_')[-1])
matches_id2
```

```python
df_1liga_20_21 = pd.DataFrame(index = range(len(matches_id2)),
                          columns=['home_team','away_team','result','goals_h','goals_a','ball_poss_h','ball_poss_a',
                                   'goal_attempts_h','goal_attempts_a','shots_on_h','shots_on_a','shots_off_h','shots_off_a',
                                   'shots_blocked_h','shots_blocked_a','free_kicks_h','free_kicks_a','corner_kicks_h','corner_kicks_a',
                                   'offsides_h','offsides_a','thorw-in_h','thorw-in_a','goalkeeper_saves_h','goalkeeper_saves_a',
                                   'fouls_h','fouls_a','yellow_h','yellow_a','red_h','red_a','passes_h','passes_a',
                                   'passes_completed_h','passes_completed_a','tackles_h','tackles_a','attacks_h','attacks_a',
                                   'attacks_dang_h','attacks_dang_a','odds_avg_h','odds_avg_a','odds_avg_d','date'])
df_1liga_20_21
```

```python
count = 0
stat_features = ['Posiadanie piłki','Sytuacje bramkowe','Strzały na bramkę','Strzały niecelne', 'Strzały zablokowane',
                 'Rzuty wolne', 'Rzuty rożne','Spalone','Wrzuty z autu', 'Interwencje bramkarzy', 'Faule','Czerwone kartki',
                 'Żółte kartki', 'Podania', 'Podania celne', 'Bloki', 'Ataki', 'Niebezpieczne ataki']
for idd in matches_id2:
    driver.get('https://www.flashscore.pl/mecz/{}/#/szczegoly-meczu/statystyki-meczu/0'.format(idd)) 
    names = driver.find_elements_by_xpath('//div[@class="participant__participantName participant__overflow"]')
    home_t = names[0].text
    away_t = names[1].text
    if (home_t not in ['Bruk-Bet T.','Radomiak Radom','Górnik Łęczna']) and (away_t not in ['Bruk-Bet T.','Radomiak Radom','Górnik Łęczna']):
        continue      
    goals = driver.find_elements_by_xpath('//div[@class="detailScore__wrapper"]')
    goals = goals[0].text.split('\n-\n')
    if goals[0] > goals[1]:
        result = 'H'
    elif goals[0] < goals[1]:
        result = 'A'
    else:
        result = 'D'
    #no_round = driver.find_elements_by_xpath('//div[@class="tournamentHeader__sportContent tournamentHeader__sportNavWrapper"]')
    #no_round = no_round[0].text.split('KOLEJKA ')[1]
    stats_dict = {}
    stats = driver.find_elements_by_xpath('//div[@class="stat__row"]')
    for i in range(len(stats)):
        feature = stats[i].text.split('\n')
        stats_dict[feature[1]] = [feature[0],feature[2]]

    if "Posiadanie piłki" in stats_dict.keys():
        ball_poss = stats_dict['Posiadanie piłki']
    else:
        ball_poss = [np.NaN,np.NaN]
    if "Sytuacje bramkowe" in stats_dict.keys():
        goal_attempts = stats_dict['Sytuacje bramkowe']
    else: 
        goal_attempts = [np.NaN,np.NaN]
    if "Strzały na bramkę" in stats_dict.keys():
        shots_on = stats_dict["Strzały na bramkę"]
    else: 
        shots_on = [np.NaN,np.NaN]
    if "Strzały niecelne" in stats_dict.keys():
        shots_off = stats_dict["Strzały niecelne"]
    else: 
        shots_off = [np.NaN,np.NaN]
    if "Strzały zablokowane" in stats_dict.keys():
        shots_blocked = stats_dict["Strzały zablokowane"]
    else: 
        shots_blocked = [np.NaN,np.NaN]
    if "Rzuty wolne" in stats_dict.keys():
        free_kicks = stats_dict["Rzuty wolne"]
    else: 
        free_kicks = [np.NaN,np.NaN]
    if "Rzuty rożne" in stats_dict.keys():
        corner_kicks = stats_dict["Rzuty rożne"]
    else: 
        corner_kicks = [np.NaN,np.NaN]
    if "Spalone" in stats_dict.keys():
        offsides = stats_dict["Spalone"]
    else: 
        offsides = [np.NaN,np.NaN]
    if "Wrzuty z autu" in stats_dict.keys():
        throw = stats_dict["Wrzuty z autu"]
    else: 
        throw = [np.NaN,np.NaN]
    if "Interwencje bramkarzy" in stats_dict.keys():
        gkp_saves = stats_dict["Interwencje bramkarzy"]
    else: 
        gkp_saves = [np.NaN,np.NaN]
    if "Faule" in stats_dict.keys():
        fouls = stats_dict["Faule"]
    else: 
        fouls = [np.NaN,np.NaN]
    if "Czerwone kartki" in stats_dict.keys():
        reds = stats_dict["Czerwone kartki"]
    else: 
        reds = [np.NaN,np.NaN]
    if "Żółte kartki" in stats_dict.keys():
        yellows = stats_dict["Żółte kartki"]
    else: 
        yellows = [np.NaN,np.NaN]
    if "Podania" in stats_dict.keys():
        passes = stats_dict["Podania"]
    else: 
        passes = [np.NaN,np.NaN]
    if "Podania celne" in stats_dict.keys():
        passes_comp = stats_dict["Podania celne"]
    else: 
        passes_comp = [np.NaN,np.NaN]
    if "Bloki" in stats_dict.keys():
        tackles = stats_dict["Bloki"]
    else: 
        tackles = [np.NaN,np.NaN]
    if "Ataki" in stats_dict.keys():
        attacks = stats_dict["Ataki"]
    else: 
        attacks = [np.NaN,np.NaN]
    if "Niebezpieczne ataki" in stats_dict.keys():
        attacks_dan = stats_dict["Niebezpieczne ataki"]
    else: 
        attacks_dan = [np.NaN,np.NaN]
    
    odds = driver.find_elements_by_xpath('//div[@class="cellWrapper"]')
    n_odds = len(odds)/3
    if n_odds == 0:
        odds_h = np.NaN
        odds_a = np.NaN
        odds_d = np.NaN
    else:
        odds_h_list = []
        odds_a_list = []
        odds_d_list = []
        for num, o in enumerate(odds):
            if num%3 == 0:
                if o.text.split('\n')[1] == '-':
                    pass
                else:
                    odds_h_list.append(float(o.text.split('\n')[1]))
            elif num%3 == 1:
                if o.text.split('\n')[1] == '-':
                    pass
                else:
                    odds_d_list.append(float(o.text.split('\n')[1]))
            else:
                if o.text.split('\n')[1] == '-':
                    pass
                else:
                    odds_a_list.append(float(o.text.split('\n')[1]))
        odds_h = np.mean(odds_h_list)
        odds_a = np.mean(odds_a_list)
        odds_d = np.mean(odds_d_list)
    date = driver.find_elements_by_xpath('//div[@class="duelParticipant__startTime"]')
    date = date[0].text
    df_1liga_20_21.loc[count] = [home_t,away_t,result,goals[0],goals[1],ball_poss[0],ball_poss[1],
                              goal_attempts[0],goal_attempts[1], shots_on[0],shots_on[1],shots_off[0],shots_off[1],
                             shots_blocked[0],shots_blocked[1], free_kicks[0],free_kicks[1],corner_kicks[0],corner_kicks[1],
                              offsides[0],offsides[1],throw[0],throw[1], gkp_saves[0],gkp_saves[1], fouls[0],fouls[1],reds[0],reds[1],
                             yellows[0],yellows[1], passes[0],passes[1],passes_comp[0],passes_comp[1],tackles[0],tackles[1],
                              attacks[0],attacks[1], attacks_dan[0],attacks_dan[1], odds_h,odds_a,odds_d, date
                             ]
    count += 1
    
df_1liga_20_21
```
